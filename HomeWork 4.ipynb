```python
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as pl
from matplotlib import pyplot as plt


img = cv2.imread('sIMG_8253.jpg')
img1 = img.copy()
imgray1 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
imgray1 = cv2.GaussianBlur(imgray1,(3,3),0)

circles = cv2.HoughCircles(imgray1, cv2.HOUGH_GRADIENT, 1.2, 30, None, 200)
count=0
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        img1 =cv2.circle(img1,(i[0], i[1]), i[2], (0, 255, 0), 2)
        img1 =cv2.circle(img1, (i[0], i[1]), 2, (255,0,0), 5)
        count += 1

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img1,'circles : '+ str(count),(10,60),font,2,(0,0,0),3,cv2.LINE_AA)



img2 = cv2.imread('sIMG_8254.jpg')
img3 = img2.copy()
imgray2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
imgray2 = cv2.GaussianBlur(imgray2,(5,5),0)

circles = cv2.HoughCircles(imgray2, cv2.HOUGH_GRADIENT, 1.8, 20, None, 200)
count=0
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        img3 =cv2.circle(img3,(i[0], i[1]), i[2], (0, 255, 0), 2)
        img3 =cv2.circle(img3, (i[0], i[1]), 2, (255,0,0), 5)
        count += 1

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img3,'circles : '+ str(count),(10,60),font,2,(0,0,0),3,cv2.LINE_AA)


img4 = cv2.imread('sIMG_8255.jpg')
img5 = img4.copy()
imgray3 = cv2.cvtColor(img4,cv2.COLOR_BGR2GRAY)
imgblur3 = cv2.GaussianBlur(imgray3,(3,7),0)

circles = cv2.HoughCircles(imgblur3, cv2.HOUGH_GRADIENT, 1.8, 30, None, 200)
count=0
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        if i[2]<50:
            img5 =cv2.circle(img5,(i[0], i[1]), i[2], (0, 255, 0), 2)
            img5 =cv2.circle(img5, (i[0], i[1]), 2, (255,0,0), 5)
            count += 1


font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img5,'circles : '+ str(count),(10,60),font,2,(0,0,0),3,cv2.LINE_AA)


plt.figure(figsize=(10,10))
plt.subplot(421),plt.imshow(img)
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(322),plt.imshow(img1)
plt.title('circles'), plt.xticks([]), plt.yticks([])
plt.subplot(323),plt.imshow(img2)
plt.title('circles'), plt.xticks([]), plt.yticks([])
plt.subplot(324),plt.imshow(img3)
plt.title('circles'), plt.xticks([]), plt.yticks([])
plt.subplot(325),plt.imshow(img4)
plt.title('circles'), plt.xticks([]), plt.yticks([])
plt.subplot(326),plt.imshow(img5)
plt.title('circles'), plt.xticks([]), plt.yticks([])
plt.show()



```
